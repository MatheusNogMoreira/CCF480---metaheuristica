```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from ils import OtimizadorILS
from ils import funcao_objetivo as funcao_objetivo_ils
from vns import OtimizadorVNS
from vns import funcao_objetivo as funcao_objetivo_vns
```

```python
limites_a = [[-100, 100], [-100, 100]]
valor_ils_a = []
solucao_ils_a = []

for i in range (30):
    otimizador_ils_a = OtimizadorILS(funcao_objetivo_ils, limites_a)
    solucao= otimizador_ils_a.otimizar()
    solucao_ils_a.append(solucao)
    valor_ils_a.append(funcao_objetivo_ils(solucao))


vizinhancas_a = [1, 10, 100]
valor_vns_a = []
solucao_vns_a = []

for i in range(30):
    otimizador_vns_a = OtimizadorVNS(funcao_objetivo_vns, limites_a, vizinhancas_a)
    solucao = otimizador_vns_a.otimizar()
    solucao_vns_a.append(solucao)
    valor_vns_a.append(funcao_objetivo_vns(solucao))


```

```python
df_ils_a = pd.DataFrame({
    "Valor": valor_ils_a,
    "Solução": solucao_ils_a
})

df_vns_a = pd.DataFrame({
    "Valor": valor_vns_a,
    "Solução": solucao_vns_a
})

print(f"\nIterated Local Search: ")
print(f"Média: {df_ils_a["Valor"].mean():.4f}")
print(f"Mínimo: {df_ils_a["Valor"].min():.4f}")
print(f"Máximo: {df_ils_a["Valor"].max():.4f}")
print(f"Desvio padrão: {df_ils_a["Valor"].std():.4f}")
print(f"Melhor solução: {df_ils_a.loc[df_ils_a["Valor"].idxmax()]["Solução"]}")

print(f"\nVariable Neighborhood Search: ")
print(f"Média: {df_vns_a["Valor"].mean():.4f}")
print(f"Mínimo: {df_vns_a["Valor"].min():.4f}")
print(f"Máximo: {df_vns_a["Valor"].max():.4f}")
print(f"Desvio padrão: {df_vns_a["Valor"].std():.4f}")
print(f"Melhor solução: {df_vns_a.loc[df_vns_a["Valor"].idxmax()]["Solução"]}")

```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize =(15,6)) 

sns.boxplot(data =df_ils_a, x = 'Valor', ax = ax1)
ax1.set_title("Iterated Local Search")


sns.boxplot(data =df_vns_a, x = 'Valor', ax = ax2)
ax2.set_title("Variable Neighborhood Search")


plt.tight_layout()
plt.show()
```
